```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

yahoo_df = pd.read_csv('data/stock_data/yahoo.csv', parse_dates=['date'], index_col=0)
yahoo_df.head()
```

```python

# Create empty dataframe to store prices across each year
prices_df = pd.DataFrame()
```

```python

# Select data for Yahoo's 2013 2014, 2015 years and concat to prices_df
for i, year in enumerate(['2013', '2014', '2015']):
    price_per_year = yahoo_df.loc[year, 'price'].reset_index()
    price_per_year.rename(columns={'price': year}, inplace=True)
    prices_df = pd.concat([prices_df, price_per_year], axis=1)
```

```python

# Plot each year's time series
prices_df[['2013', '2014', '2015']].plot()
plt.show()
```
