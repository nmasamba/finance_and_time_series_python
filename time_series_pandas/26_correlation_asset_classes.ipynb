```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Correlation is a measure of the LINEAR relationship between two variables
# In financial markets, correlation is importantant for predictive modelling and risk management
assets_df = pd.read_csv('data/stock_data/asset_classes.csv', parse_dates=['DATE'], index_col='DATE')
assets_df.head()
```

```python
assets_df.info()
```

```python

daily_returns = assets_df.pct_change()
daily_returns.head()
```

```python

# Scatter plot shows relationship among any two variables
sns.jointplot(x='SP500', y='Gold', data=daily_returns)
plt.show()
```

```python

# Calculate all correlations
correlations = daily_returns.corr()
print(correlations)
```

```python

# Visualise all correlations
sns.heatmap(correlations, annot=True)
plt.show()
```
