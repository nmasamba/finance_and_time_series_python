```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Note: Co2 data not sampled on some days (e.g. no record for 2 Jan)
co_df = pd.read_csv('data/air_quality_data/co_cities.csv', parse_dates=['date'], index_col=0)
co_df.head(10)
```

```python
co_df.info()
```

```python

# Change sampling frequency to daily(D) - downsample
# This creates NaN observations to account for days on which no samples were taken = missing data
co_df_daily = co_df.asfreq('D')
co_df_daily.head(10)
```

```python
co_df_daily.info()
```

```python
co_df_daily.plot(subplots=True)
plt.show()
```

```python

# Change sampling frequency to monthly(M) - upsample
# Data cascaded to represent monthly intervals
co_df_monthly = co_df.asfreq('M')
co_df_monthly.head(10)
```

```python
co_df_monthly.info()
```

```python
co_df_monthly.plot(subplots=True)
plt.show()
```
