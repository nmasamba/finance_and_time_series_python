```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

ozone_df = pd.read_csv('data/air_quality_data/ozone_nyla.csv', parse_dates=['date'], index_col='date')
ozone_df.head()
```

```python
ozone_df.info()
```

```python

# Since there's no frequency data, we can assign it without modifying the underlying df useing asfreq()
ozone_df.asfreq('D').info()
```

```python

# We'll use resample() to assign and convert frequencies and plot the result
# Downsample to Weekly frequency and aggregate as mean
ozone_df.resample('W').mean().plot()
plt.show()
```

```python

# Monthly frequency
ozone_df.resample('M').mean().plot()
plt.show()
```

```python

# Annual frequency
ozone_df.resample('A').mean().plot()
plt.show()
```
