```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('data/stock_data/sp500.csv', parse_dates=['date'], index_col='date')
data.head()
```

```python
#Â Compute running min and max
data['running_min'] = data['SP500'].expanding().min()
data['running_max'] = data['SP500'].expanding().max()
data.head()
```

```python

# Plot running stats with time series
data.plot()
plt.show()
```
