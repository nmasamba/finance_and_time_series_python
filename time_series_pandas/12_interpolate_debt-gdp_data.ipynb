```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load US debt/GDP + unemployment data
# Note that Debt/GDP is sampled quarterly while Unemployment is sampled monthly
data = pd.read_csv('data/stock_data/debt_unemployment.csv', parse_dates=['date'], index_col='date')
data.head()
```

```python

# Debt/GDP not plotted due to NaNs
data.plot(secondary_y='Unemployment')
plt.show()
```

```python

# Interpolate the entire dataframe
# Pandas automatically interpolates missing Debt/GDP values according to higher frequency
interpolated = data.interpolate()
interpolated.head()
```

```python

# Plot interpolated data
interpolated.plot(secondary_y='Unemployment')
plt.show()
```
