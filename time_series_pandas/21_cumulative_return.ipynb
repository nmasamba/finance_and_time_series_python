```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

sp500 = pd.read_csv('data/stock_data/sp500.csv', parse_dates=['date'], index_col='date')
sp500.head()
```

##Â Calculating a multi-period return

Single period return: $r_t = \frac{P_t}{P_{t-1}} - 1$ </br>

Multi-period return: $R_T = (1+r_1)(1+r_2) ... (1+r_T) - 1$ </br> 

```python

single_period_ret = sp500['SP500'].pct_change()
multi_period_ret = single_period_ret.add(1).cumprod().sub(1)
```

```python

# Plot multi-period return %
multi_period_ret.mul(100).plot()
plt.show()
```
