```python

import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv('/Users/nyashamasamba/Downloads/BTC-USD.csv', parse_dates=['Date'])
```

```python
prices.head()
```

```python
# Sort prices data by parsed date
prices = prices.sort_values(by='Date')
prices.set_index('Date', inplace=True)
```

```python
prices.head()
```

```python
# Calculate percentage simple return
# Note that first day will be nan 
prices['SimpleReturns'] = prices['Adj Close'].pct_change()
```

```python
prices
```

```python
# Plot frequency distribution of returns
plt.hist(prices.SimpleReturns.dropna(), bins=math.ceil(np.sqrt(len(prices.SimpleReturns.dropna()))), density=False)
plt.show()
```

```python
# Plot returns as time series
prices.SimpleReturns.plot()
plt.show()
```
